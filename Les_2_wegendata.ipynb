# Overzicht van de straten

We gaan nu folium laden, daarmee kun je landkaarten laden en punten weergeven. op een kaart. Dan gaan we alle punten weergeven. Sommige punten maken we een andere kleur, op basis van een list van 5 elementen.

````python
import folium

# Create a folium map centered at a specific location
map_center = [52.3702, 4.8952]  # Amsterdam coordinates
zoom_level = 15
map_amsterdam = folium.Map(location=map_center, zoom_start=zoom_level)

# Iterate over the rows in the DataFrame
def maak_een_kaart(df, gemarkeerde_straten, gemarkeerde_straten_tuple):
  for _, row in df.iterrows():
      # Extract the coordinates from the JSON data
      coordinates = (row[0]['geometry']['coordinates'][0][0])
      straatnaam = (row['features']['properties']['street_name'])

      # Convert coordinate pairs to string values
      coordinate_pairs = []
      for coord in coordinates:
        coordinate_pairs.append(str(coord))
      reversed_coordinate_pair = coordinate_pairs[::-1]

      # Voor iedere locatie wordt een marker gemaakt. Op basis van de straatnaam en de lijst/tuple die je eerder gemaakt hebt, krijgen ze een andere kleur.
      ## let op: er zitten fouten in!! ;-)
      if straatnaam in gemarkeerde_straten
        marker = folium.Marker(location=reversed_coordinate_pair, icon=folium.Icon(color='green', tooltip=straatnaam))
      elif straatnaam in gemarkeerde_straten_tuple:
      marker = folium.Marker(location=reversed_coordinate_pair, icon=folium.Icon(color='bla', tooltip=straatnaam))
      else:
        marker = folium.Marker(location=reversed_coordinate_pair, icon=folium.Icon(color='orange', tooltip=straatnaam))

      # Als de kleur gekozen is, wordt hij aan de kaart toegevoegd.
      marker.add_to(map_amsterdam)

  # Display the map
  # return map_amsterdam
````

````python
def get_traffic_sign_events(offset=None, limit=None, town_code=None):
    base_url = "https://data.ndw.nu/api/rest/static-road-data/traffic-signs/v2/events"
    params = {}
    if offset:
        params['offset'] = offset
    if limit:
        params['limit'] = limit
    if town_code:
        params['town-code'] = town_code

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        # Create a DataFrame from the JSON data
        df = pd.DataFrame(data)
        return response.json()
    else:
        return None

def get_traffic_sign_current_state(town_code=None):
    base_url = "https://data.ndw.nu/api/rest/static-road-data/traffic-signs/v2/current-state"
    params = {}
    if town_code:
        params['town-code'] = town_code

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        # Create a DataFrame from the JSON data
        df = pd.DataFrame(data)
        return response.json()
    else:
        return None
````
