## While true
Combinatie van een loop met een if statement. De loop blijft herhalen zolang de if statement waar is.

````python
## Een paar voorbeelden van for.. en while... loops

studenten = {"Piet": 7, "Klaas": 8, "Jan": 9, "Kees": 10}
for student in studenten:
    print(student + " heeft een " + str(studenten[student]))

for i in range(3, 10):
    print("Dit is de " + str(i) + "e keer dat ik dit print")

i=1
while i < 10:
    print(f"Dit is de {i}e keer dat de loop wordt uitgevoerd") #met een f tussen de hakkjes kun je een variabele in een string zetten{}
    i = i + 1
````

````python
# Provided JSON data
pizza_order = {
    "salami": {
        "type": "pizza",
        "price": 12.95
    },
    "kaas": {
        "type": "pizza",
        "price": 10.95
    },
    "aubergine": {
        "type": "pizza",
        "price": 8.95
    },
    "toppings": [
        {"type": "extra kaas", "price": 2.00},
        {"type": "chanpignons", "price": 1.50}
    ],
    "drinks": [
        {"type": "cola", "size": "medium", "price": 2.00}
    ],
    "sides": [
        {"type": "frites", "size": "small", "price": 3.50}
    ]
}

# De kosten voor de order
totaalbedrag=0


# Deze loop loopt door zolang er geen break gegeven wordt
while True
    # Vraag de bestelling
order = input("Vul je bestelling in (type 'exit' om te stoppen): ")

    # Stop als er exit getypt wordt
    if order.lower() == 'exit':
        break

    # Als er meer items zijn met een komma ertussen, wordt ieder item los meegenomen
    items = order.split(',')

    # Beslis voor ieder item:
    for item in items:
        # Hiermee haal je spaties aan het begin of eind weg. python ziet de string "pizza " anders dan "pizza"
        item = item.strip()

        Kijk of het een pizza is
        if item in pizza_order:
            totaalbedrag += pizza_order[item]["price"]
            print(f"De {item} pizza zit in je bestelling.")
        # Check if the item is a topping
        elif any(topping["type"] == item for topping in pizza_order["toppings"]):
            totaalbedrag += next(topping["price"] for topping in pizza_order["toppings"] if topping["type"] == item)
            print(f"Added {item} topping to your order.")
        # Check if the item is a drink
        elif any(drink["type"] == item for drink in pizza_order["drinks"]):
            totaalbedrag += next(drink["price"] for drink in pizza_order["drinks"] if drink["type"] == item)
            print(f"{item} is toegevoegd.")
        # Check if the item is a side
        elif any(side["type"] == item for side in pizza_order["sides"]):
            totaalbedrag += next(side["price"] for side in pizza_order["sides"] if side["type"] == item)
            print(f"{item} toegevoegd aan je bestelling.")
        else:
            print(f"Sorry, {item} Kun je niet bestellen.")

print(f"Het totaalbedrag van je bestelling is Eur {totaalbedrag:.2f}")
````
